```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("iris")
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.histplot(data=df,x="species")
```

We have equal sample size of all three species

```python
sns.kdeplot(data=df,x="sepal_length")
```

Most sepal length between 4.5 and 7 units

```python
sns.kdeplot(data=df,x="sepal_length",hue ="species")
```

Virginica species generally has longest sepal length and setosa has shortest. 

```python
sns.kdeplot(data=df,x="sepal_width")
```

Most sepal width are are between 2.5 and 3.5

```python
sns.kdeplot(data=df,x="sepal_width",hue ="species")
```

It would be impossible to tell the species with sepal width, but under some cicumstances it is possible. If sepal width is greater than 4.7-4.8 it is very likely to be setosa and if it is lower than 1.7-1.8 it is very likely to be versicolour.

```python
sns.kdeplot(data=df,x="petal_length")
```

Petal length is very up and down. No clear portion where it is the greatest, however most petal length is between 0.2-2.7 and between 3.5-6.5.

Petal

```python
sns.kdeplot(data=df,x="petal_length",hue ="species")
```

Petal length instantly tells us whether or not the plant is of setosa species. It is more difficult to differentiate between versicolour and virginica if petal length is between 3.8 and 5.8. Otherwise we can easily tell what species the plant belongs to. Petal length is very important to identify the species of plant.

```python
sns.kdeplot(data=df,x="petal_width")
```

very simmilar to petal length. Petal width is very up and down. No clear portion where it is the greatest, however most petal width is between 0.1-0.8 and between 1-2.5.

```python
sns.kdeplot(data=df,x="petal_width",hue ="species")
```

Very simmilar to petal length. Petal width instantly tells us whether or not the plant is of setosa species. It is more difficult to differentiate between versicolour and virginica if petal width is between 1 and 2. Otherwise we can easily tell what species the plant belongs to. Petal width is very important to identify the species of plant.

```python
sns.scatterplot(data=df,x="petal_width",y ="petal_length")
```

there is a positive correlation between petal length and width.
